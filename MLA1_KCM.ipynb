## 9. Write the code to add a column named temp_f to the dataset that contains the temperature from temp_c represented in Fahrenheit.

```python
#adding Fahrenheit column to the dataset
#formula for convesion (F = °C × (9/5) + 32)
df['temp_f']= df['temp_c'] * (9/5) +32
#checking that the changes were correct
print(f"\nFahrenheit column head\n{df['temp_f'].head()}\n")
print(f"Celcius column head\n{df['temp_c'].head()}\n")
```

## 12. The date_time column is not actually considered a date by Pandas.  Write the code to convert this column to a date/time column and to verify the conversion was successful.

```python
#changing the date time column to be datetime in pandas
#date time conversion
import datetime
df['date_time'] = pd.to_datetime(df['date_time'])
#validating that system sees column in date_time
print(f"Validating date time change in system: {df['date_time'].dtype}\n")
```

## 13. Write the code to get the average temperature (in F) for each month and write it into an object called monthly_avg. This object should be a Pandas Series which is similar to a Pandas Dataframe except it only has one column

```python
#getting new dataframe of average months temps in F
month = df['date_time'].dt.month
monthly_avg = df['temp_f'].groupby(month).mean()
print(f"Monthly average temperature in F of each month in a data frame\n {monthly_avg}\n")
```

## 14. Write code to generate a scatterplot that compares the temperature against the relative humidity. The size of your plot should 20 x 10.

```python
#Creating a scatter plot that compares the temperature against the relative humidity.
#setting figure size to 20x10
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['temp_f'],y=df['relative_humidity'], color='green')
#giving a title and naming x and y axis
plt.title('Temperature vs Relative Humidity')
plt.xlabel('Temperature (F)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

## 15. Write the code to generate a line chart that shows the average temperature per month. The size of your plot should 20 x 10.

```python
# Write the code to generate a line chart that shows the average temperature per month. The size of your plot should 20 x 10 and should look like the plot below
#setting plot size
plt.figure(figsize=(20,10))
#generating plot
sns.lineplot(x=monthly_avg.index, y=monthly_avg.values, color='red')
#naming the lineplot
plt.title('Average Temperature per Month')
plt.xlabel('Month')
plt.ylabel('Temperature (F)')
plt.show()
```

## 16. Write the code to generate a box plot. The size of your plot should 20 x 10.

```python
# Write the code to generate a box plot. The size of your plot should 20 x 10 and should look like the plot below
#Setting figure size
plt.figure(figsize=(20,10))
#generating boxplot
plot_data=df[['temp_f','temp_c','dew_point','relative_humidity','wind_speed','visibility','pressure']]
plot_data.boxplot()
#removing the grid
plt.grid(False)
plt.show()
```
